## Step 5: Create any additional visualizations that will help to answer the question(s) you want to answer.

```python
# I will now create a heat map visualization to show the 
# differences and similarities between the different variables
plt.figure(figsize=(15,10))
sns.heatmap(Netflix_Data.corr(), annot=True);
```

```python
# I will now create a bar chart for the top 10 cost for the 
# basic subscription with the use of matplotlib.pyplot.
plt.figure(figsize = (12,7))
top_10 = Netflix_Data['Country'][:10]
chart = Netflix_Data.groupby(top_10)['Cost Per Month - Basic ($)'].sum().sort_values(ascending = False).plot(kind = 'bar', 
                                                                                                  color = 'green')
chart.set_xticklabels(chart.get_xticklabels(), rotation = 45)
plt.title('Top 10 of Country based on Total Basic Cost Per Month', fontsize = 15, pad = 12)
plt.xlabel('Country', fontsize = 12)
plt.ylabel('Cost Per Month - Basic ($)', fontsize = 12)
plt.show()
```

As seen above we see that the country with the highest cost per month for the basic subscription is belgium at a cost of about 10 dollars and the lowest cost of the top 10 is argentina which is right under 4 dollars a month.
